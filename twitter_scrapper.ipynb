```python
set_post  = list(dict.fromkeys(post_urls))

for link in set_post:
    flag=True
    check_cnt=0
    driver.get(link)
    time.sleep(3)
    
    attempt=0
    replies = driver.find_elements(By.XPATH,"//article[@data-testid='tweet'][@tabindex='0']")
    
    while flag==True:
        last_position = driver.execute_script("return window.pageYOffset;")
        for reply in replies:
            
            user = driver.find_element(By.XPATH,".//article[@data-testid='tweet'][@tabindex='0']//div[@data-testid='User-Names']//a")
            commenter =  user.text
            profile_url = user.get_attribute('href')
            comment_time = driver.find_element(By.XPATH,".//article[@data-testid='tweet'][@tabindex='0']//time").get_attribute('datetime')
            
            comment = driver.find_element(By.XPATH,".//article[@data-testid='tweet'][@tabindex='0']//div[@data-testid='tweetText']").text
            
            
            tweet_link = link
            commenters.append(commenter)
            profile_urls.append(profile_url)
            comments.append(comment)
            comment_times.append(comment_time)
            tweet_links.append(tweet_link)
            print(comment[:10],"~",end=' ')
            comment=""


        js_scroll = "window.scrollBy(0, 200);"
        driver.execute_script(js_scroll)
        time.sleep(2)
        replies = driver.find_elements(By.XPATH,"//article[@data-testid='tweet'][@tabindex='0']")
        current_position = driver.execute_script("return window.pageYOffset;")
        print(attempt,end=' ')
        if last_position == current_position:
            try:
                more = driver.find_element(By.XPATH,"//span[contains(text(),'Show more replies')]")
                more.click()
            except:
                attempt += 1
            #counting failed attempt of scrolling
            if attempt >= 4:
                flag = False
                break
            else:
                time.sleep(2) 

    
        
    
```

## Quoted retweet

```python
users_retweeted=[]
profile_urls=[]
quotes = []
tweet_links = []
```

```python
set_post  = list(dict.fromkeys(post_urls))

for link in set_post:
    flag=True
    check_cnt=0
    driver.get(link+"/retweets/with_comments")
    time.sleep(3)
    
    attempt=0
    try:
        empty_header = driver.find_elements(By.XPATH,"//div[@data-testid='empty_state_header_text']/span[contains(text(),'No Quote Tweets yet')]")
        if empty_header:
            print("No retweets")
            continue
    except:
        pass
    quoted_retweets = driver.find_elements(By.XPATH,"//article[@data-testid='tweet'][@tabindex='0']")
    while flag==True:
        last_position = driver.execute_script("return window.pageYOffset;")
        for retweet in quoted_retweets:
            
            user = driver.find_element(By.XPATH,".//article[@data-testid='tweet'][@tabindex='0']//div[@data-testid='User-Names']//a")
            user_retweeted =  user.text
            profile_url = user.get_attribute('href')
            try:
                quote = driver.find_element(By.XPATH,".//article[@data-testid='tweet'][@tabindex='0']/div/div/div/div[2]/div[2]/div[2]/div[1]/div[@data-testid='tweetText']").text
            
            except:
                quote = ""
            tweet_link = link
            users_retweeted.append(user_retweeted)
            profile_urls.append(profile_url)
            quotes.append(quote)
            tweet_links.append(tweet_link)
            print(quote[:10],"~",end=' ')
            quote=""


        js_scroll = "window.scrollBy(0, 200);"
        driver.execute_script(js_scroll)
        time.sleep(3)
        quoted_retweets = driver.find_elements(By.XPATH,"//article[@data-testid='tweet'][@tabindex='0']")
        current_position = driver.execute_script("return window.pageYOffset;")
        print(attempt,end=' ')
        if last_position == current_position:
            attempt += 1
            #counting failed attempt of scrolling
            if attempt >= 2:
                flag = False
                break
            else:
                time.sleep(2) 

    
        
    
```

```python
quote_retweet_df = pd.DataFrame(zip(tweet_links,users_retweeted,profile_urls,quotes)
                  ,columns=['tweet_links','users_retweeted','profile_urls','quotes'])
quote_retweet_df=quote_retweet_df.drop_duplicates()
quote_retweet_df
```
